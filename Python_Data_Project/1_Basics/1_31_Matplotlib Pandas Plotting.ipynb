```python
!pip install datasets
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## ðŸŸ© Job Posting Location Bar Plot (1.31.1)

```python
job_platform = df['job_via'].value_counts().head()
job_platform.plot(kind='bar')
plt.show()
```

---

## ðŸŸ© Job Postings Trend Line Plot (1.31.2)

```python
monthly_job_counts = df['job_posted_date'].dt.to_period('M').value_counts().sort_index()
monthly_job_counts.plot(kind='line')
plt.xlabel('Month')
plt.ylabel('Number of Job Postings')
plt.title('Job Postings Trend')
plt.show()
```

---

## ðŸŸ¨ Top Job Locations (1.31.3)

```python
job_location_count = df['job_location'].value_counts().head()
job_location_count.plot(kind='barh')
plt.xlabel('job count')
plt.ylabel('Job Location')
plt.title('Number of Job Postings by Job Location')
plt.show()
```
