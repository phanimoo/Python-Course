```python
!pip install datasets
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## üü© Bar Chart with Labels (1.30.1)

```python
job_country_counts = df['job_country'].value_counts().head()
plt.bar(job_country_counts.index, job_country_counts)
plt.xlabel('Country')
plt.ylabel('Number of Postings')
plt.title('Number of Jobs by Country')
plt.xticks(rotation=45, ha='right')
plt.show()
```

---

## üü® Horizontal Bar Chart with Labels (1.30.2)

```python
job_salary = df.groupby('job_title_short')['salary_year_avg'].mean()
job_salary = job_salary.sort_values(ascending=True)

plt.barh(job_salary.index, job_salary.values)
plt.xlabel('Average Salary')
plt.title('Average Salary by Job Title')
plt.show()
```

---

## üü• Line Chart with Labels (1.30.3)

```python
df['job_posted_month'] = df['job_posted_date'].dt.month
job_month_salary = df.groupby(by='job_posted_month')['salary_year_avg'].mean().sort_index()

plt.plot(job_month_salary.index, job_month_salary, color="red", label='Average Salary')
plt.xlabel('Month')
plt.ylabel('Average Salary')
plt.title('Average Salary by Month')
plt.legend()
plt.show()
```
