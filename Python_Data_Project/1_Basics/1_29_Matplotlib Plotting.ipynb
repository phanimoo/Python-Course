```python
!pip install datasets
import pandas as pd
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## 🟩 Job Country Bar Chart (1.29.1)

```python
import matplotlib.pyplot as plt
country_counts = df['job_country'].value_counts().head(5)
plt.bar(country_counts.index, country_counts.values, color='orange')
plt.show()
```

---

## 🟨 Average Salary Horizontal Bar Chart (1.29.2)

```python
job_titles = df.groupby(by='job_title_short')['salary_year_avg'].mean().sort_values(ascending=True)
plt.barh(job_titles.index, job_titles.values)
plt.show()

```

---

## 🟥 Average Salary Trend Line Plot (1.29.3)

```python
job_month = df['job_posted_date'].dt.month
job_month_avg = df.groupby(by=job_month)['salary_year_avg'].mean()
plt.plot(job_month_avg.index, job_month_avg.values)
plt.show()
```
