## ðŸŸ¨ Filter Job Postings by Location (1.18.2)

```python
%%writefile job_filter.py
# cell magic command to write the contents of this cell to a file called job_filter.py
# function filter_by_location that takes a list of job postings and a location and
# returns a list of job postings in that location.
# Import this module and use the function to filter job postings in 'New York'.
# Use the variable of job_postings shown below.
def filter_by_location(job_postings, location):
    # filtered_postings = []
    # for job in job_postings:
    #     if job['location'] == location:
    #         filtered_postings.append(job)
    # filtered_postings = [job for job in job_postings if job['location'] == location]
    filtered_postings = list(filter(lambda job: job['location'] == location, job_postings))

    return filtered_postings
```

```python
# Code Needed
import job_filter
job_postings = [
    {'title': 'Data Scientist', 'location': 'New York'},
    {'title': 'Data Analyst', 'location': 'San Francisco'},
    {'title': 'Machine Learning Engineer', 'location': 'New York'}
]

job_filter.filter_by_location(job_postings, 'New York')
```
