```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## ðŸŸ© Concat Job Postings (2.8.1)

```python
df_data_analyst = df[df['job_title'] == 'Data Analyst'].copy()
df_data_scientist = df[df['job_title'] == 'Data Scientist'].copy()
df_combined = pd.concat([df_data_analyst, df_data_scientist],ignore_index=True)
df_combined
```

---

## ðŸŸ¨ Concat Monthly Postings (2.8.2)

```python
df_oct = df[df['job_posted_date'].dt.month == 10].copy()
df_nov = df[df['job_posted_date'].dt.month == 11].copy()
df_dec = df[df['job_posted_date'].dt.month == 12].copy()
df_combined_q4 = pd.concat([df_oct, df_nov, df_dec], ignore_index=True)
print(df_combined_q4)
df_combined_q4['job_posted_date'].dt.strftime('%b').value_counts().plot(kind='bar')
```

---

## ðŸŸ¨ Concat Job Types (2.8.3)

```python
df_full_time = df[df['job_schedule_type'] == 'Full-time'].copy()
df_part_time = df[df['job_schedule_type'] == 'Part-time'].copy()
df_combined_schedule = pd.concat([df_full_time, df_part_time], ignore_index=True)
df_combined_schedule['job_schedule_type'].value_counts().plot(kind='bar')
```
