```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

## ðŸŸ© Health Insurance Pie Chart (2.14.1)

```python
df_DA = df[df['job_title'] == 'Data Analyst'].copy()
health_insurance_counts = df_DA['job_health_insurance'].value_counts()

plt.pie(health_insurance_counts, labels=['No', 'Yes'], startangle=90, autopct='%1.1f%%')
plt.title('Proportion of Data Analyst Jobs Mentioning Health Insurance')

# Display the chart
plt.show()
```

---

## ðŸŸ¨ Job Schedule Distribution (2.14.2)

```python
df_ft = df[df['job_schedule_type'] == 'Full-time'].copy()
df_pt = df[df['job_schedule_type'] == 'Part-time'].copy()
df_ft_pt = pd.concat([df_ft, df_pt],ignore_index=True)
df_ft_pt = df_ft_pt['job_schedule_type'].value_counts()
df_ft_pt.plot(kind='pie', labels=df_ft_pt.index, autopct='%11.1f%%', startangle=90, legend=False)
plt.title('Job Schedule Distribution Full-Time vs. Part-Time')
```

---

## ðŸŸ¥ Degree Mentioned Pie Chart (2.14.3)

```python
df_US = df[df['job_country'] == 'United States'].copy()
df_US_no_degree = df_US['job_no_degree_mention'].value_counts()
df_US_no_degree.plot(kind='pie', labels=['Degree Mentioned', 'No Degree Mentioned'], autopct='%11.1f%%', startangle=90, legend=False, explode=(0, 0.1))
plt.title('Proportion of Jobs in the United States Mentioning No Degree')
plt.ylabel(' ')
plt.show()
```
