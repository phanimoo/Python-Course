```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## ðŸŸ© Merge US Jobs and Monthly Postings (2.7.1)

```python
df['job_posted_month'] = df['job_posted_date'].dt.to_period('M')
df_usa = df.loc[df['job_country'] == 'United States'].copy()
df_usa
df_month = df.groupby('job_posted_month').size().reset_index(name='postings_count').copy()
df_merge = df_usa.merge(df_month, on='job_posted_month', how='right')
df_merge.head()
```

---

## ðŸŸ¨ Merge Non-Remote Jobs and Salary (2.7.2)

```python
df_not_remote = df.loc[df['job_work_from_home'] == False].copy()
df_not_remote_salary = df_not_remote.groupby('job_title_short')['salary_year_avg'].mean().reset_index(name='salary_not_remote_avg')
df_merge_salary = df_not_remote.merge(df_not_remote_salary, on='job_title_short', how='right')
df_merge_salary.head()
```

---

## ðŸŸ¨ Merge Company Salary and Remote Count (2.7.3)

```python
df_company_salary = df.groupby('company_name')['salary_year_avg'].mean().reset_index().copy()
df_company_remote = df[df['job_work_from_home']==True].groupby('company_name').size().reset_index(name='company_remote_count').copy()
df_company_merge = df_company_salary.merge(df_company_remote, on='company_name')
df_company_merge.head()
```
