```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

## ðŸŸ© Data Scientist Hourly Salaries (2.17.1)

```python
df_DS = df[df['job_title_short'] == 'Data Scientist'].copy()
df_DS['salary_hour_avg'].plot(kind='hist', bins=40, color='blue', edgecolor='black', figsize=(10,7))
plt.title('Distribution of Data Scientist Hourly Salary')
plt.xlabel('Salary')
plt.ylabel('Number of Jobs')
plt.show()
```

---

## ðŸŸ¨ Job Postings by Number of Skills (2.17.2)

```python
df_skills = df[df['job_skills'].notna()].copy()
df_skills['num_skills'] = df_skills['job_skills'].apply(len)
df_skills['num_skills'].plot(kind='hist', bins=55, color='magenta', edgecolor='black', figsize=(10,7))
plt.title('Distribution of Number of Skills Required')
plt.xlabel('Number of Skills')
plt.ylabel('Number of Jobs')
plt.show()
```

---

## ðŸŸ¥ Data Engineer Salaries (2.17.3)

```python
df_DE = df[df['job_title_short'] == 'Data Engineer'].copy()
df_DE['salary_year_avg'].plot(kind='hist', bins=50, color='red', edgecolor='black', figsize=(10,7))
mean_salary = df_DE['salary_year_avg'].mean()
median_salary = df_DE['salary_year_avg'].median()
mean_line = plt.axvline(mean_salary, label=f'Mean: ${mean_salary}', color='blue', linestyle='dashed', linewidth=1)
median_line = plt.axvline(median_salary, label=f'Median: ${median_salary}', color='green', linestyle='dashed', linewidth=1)
plt.title('Distribution of Data Engineer Yearly Salary')
plt.xlabel('Salary')
plt.ylabel('Number of Jobs')
plt.legend(handles=[mean_line, median_line])
plt.show()
```
