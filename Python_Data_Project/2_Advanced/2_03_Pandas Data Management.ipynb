```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## ðŸŸ© Drop Missing Values (2.3.1)

```python
df_altered = df.copy()
df_no_nan = df_altered.dropna(subset=['salary_year_avg'])
df_no_nan.head()
```

---

## ðŸŸ© Random Sample of DataFrame (2.3.2)

```python
df_sampled = df.copy()
df_sampled = df_sampled.sample(frac=0.1, random_state=1)
df_sampled.head()
```

---

## ðŸŸ¨ Fill Missing Salaries (2.3.3)

```python
df_altered = df.copy()
df_altered['salary_year_avg'] = df_altered['salary_year_avg'].fillna(df_altered['salary_year_avg'].median())
df_altered.head()
```
