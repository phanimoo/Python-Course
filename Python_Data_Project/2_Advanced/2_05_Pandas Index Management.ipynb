```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## 🟩 Set Index and Display Rows (2.5.1)

```python
df_date = df.set_index('job_posted_date').copy()
df_date.head()
```

---

## 🟩 Reset Index and Drop Existing (2.5.2)

```python
df_date = df_date.reset_index()
df_date
```

---

## 🟩 Sort Index (2.5.3)

```python
df_median = df.pivot_table(index = 'company_name', values = 'salary_year_avg', aggfunc= 'median').copy()
df_median = df_median.sort_index(ascending = False)
df_median.head()
```

---

## 🟨 Pivot Table Reset & Set Index (2.5.4)

```python
df_job_median = df.pivot_table(index = 'job_title_short', values = 'salary_year_avg', aggfunc= 'median').copy()
df_job_median = df_job_median.reset_index()
df_job_median = df_job_median.set_index('salary_year_avg')
df_job_median.head()
```
