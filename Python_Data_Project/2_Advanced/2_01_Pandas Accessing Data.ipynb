```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## ðŸŸ© Select First 5 Rows with .iloc() (2.1.1)

```python
df.iloc[0:5, 1:3]
```

---

## ðŸŸ© Get Rows and Columns with .loc() (2.1.2)

```python
df.loc[5:9, 'job_title':'job_posted_date']
```

---

## ðŸŸ¨ First 15 Rows in US (2.1.3)

```python
filtered_df = df[df['job_country'] == 'United States']
filtered_df.iloc[0:15, [1, 12]]
```

---

## ðŸŸ¥ Filter by Date (2.1.4)

```python
df_dates = df[df['job_posted_date'] > '2023-05-01']
df_dates.loc[:, ['job_title', 'job_location', 'job_posted_date']]
```
