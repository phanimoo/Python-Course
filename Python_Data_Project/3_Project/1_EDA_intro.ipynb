```python
# Importing Libraries
import ast
import pandas as pd
import seaborn as sns
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

## Filter for Data Analyst roles in the US

```python
df_DA_US = df[(df['job_title_short'] == 'Data Analyst') & (df['job_country'] == 'United States')].copy()
```

```python
df_plot = df_DA_US['job_location'].value_counts().head(10).to_frame()
sns.set_theme(style='ticks')
sns.barplot(data=df_plot, x=df_plot['count'], y='job_location', hue='count', palette='dark:b_r')
plt.title('Top 10 Locations for Data Analyst Jobs in the US')
plt.xlabel('Number of Jobs')
plt.ylabel(' ')
plt.legend().set_visible(False)
plt.show()
```

```python
# rewrite the above with a for loop
dict_column = {
    'job_work_from_home': 'Work from Home Offered',
    'job_no_degree_mention': 'Degree Requirement',
    'job_health_insurance': 'Health Insurance Offered'
}

fig, ax = plt.subplots(1, 3)
fig.set_size_inches((12, 5))

for i, (column, title) in enumerate(dict_column.items()):
    ax[i].pie(df_DA_US[column].value_counts(), labels=['False', 'True'], autopct='%1.1f%%', startangle=90)
    ax[i].set_title(title)

# plt.suptitle('Benefit Analysis of Data Jobs', fontsize=16)
plt.show()
```

```python
df_plot = df_DA_US['company_name'].value_counts().head(10).to_frame()
sns.set_theme(style='ticks')
sns.barplot(data=df_plot, x=df_plot['count'], y='company_name', hue='count', palette='dark:b_r')
plt.title('Top 10 Companies for Data Analyst Jobs in the US')
plt.xlabel('Number of Jobs')
plt.ylabel(' ')
plt.legend().set_visible(False)
plt.show()
```

```python
#df_DA_US['salary_year_avg'].plot(kind='hist', bins=50, color='red', edgecolor='black', figsize=(10,7))
sns.set_theme(style='ticks')
sns.displot(df_DA_US['salary_year_avg'], kind='hist', bins = 25, color='darkred', edgecolor='black')
mean_salary = df_DA_US['salary_year_avg'].mean()
median_salary = df_DA_US['salary_year_avg'].median()
mean_line = plt.axvline(mean_salary, label=f'Mean: ${mean_salary:.2f}', color='blue', linestyle='dashed', linewidth=1)
median_line = plt.axvline(median_salary, label=f'Median: ${median_salary:.2f}', color='green', linestyle='dashed', linewidth=1)
plt.title('Distribution of Data Analyst Yearly Salary in the US')
plt.xlabel('Salary')
plt.ylabel('Number of Jobs')
plt.xlim(0, 200000)
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos: f'${int(x/1000)}k'))
plt.legend(handles=[mean_line, median_line])
plt.show()
```
